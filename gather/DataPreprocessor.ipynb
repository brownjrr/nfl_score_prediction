```python
import pandas as pd
import os
```

# Cleaning Players DataFrame

```python
players_df = pd.read_csv("../data/players.csv")

# remove any "Unnamed" columns
players_df = players_df.drop([i for i in players_df.columns if i.startswith("Unnamed:")], axis=1)

# adding id column
players_df['player_id'] = players_df['link'].str.split('/', expand=False).str[-1].str.replace(".htm", "")

assert players_df['player_id'].shape[0] == players_df['player_id'].drop_duplicates().shape[0]

# reordering columns
id_cols = ['player_id']
players_df = players_df[id_cols + [i for i in players_df.columns if i not in id_cols]]

players_df.to_csv("../data/players.csv", index=False)

players_df
```

# Cleaning Rosters DataFrame

```python
rosters_df = pd.read_csv("../data/rosters.csv")

# create player_id columns based on player_link
rosters_df['player_id'] = rosters_df['player_link'].str.split("/", expand=False).str[-1].str.replace(".htm", "")

# drop player_link column
rosters_df = rosters_df.drop(["player_link"], axis=1)

# create team_link column based on link column
rosters_df['team_link'] = rosters_df['link'].str.split("/").str[:-1].str.join("/") + '/'

# rename link to roster_link
rosters_df = rosters_df.rename(columns={'link': 'roster_link'})

# reorder columns
id_cols = ['roster_link', 'team', 'roster_year', 'team_link', 'player_id', 'player_name']
rosters_df = rosters_df[id_cols + [i for i in rosters_df.columns if i not in id_cols]]

# rosters_df.to_csv("../data/rosters.csv", index=False)

rosters_df
```

# Cleaning Temp Teams Table

```python
teams_df = pd.read_csv("../data/temp_teams.csv")

# creating team_id column
teams_df['team_id'] = teams_df['team_url'].str.split('/').str[-2]

# reordering columns
id_cols = ['team_id', 'current_team_nm', 'team_nm', 'team_url']
teams_df = teams_df[id_cols + [i for i in teams_df.columns if i not in id_cols]]

# teams_df.to_csv("../data/temp_teams.csv", index=False)

teams_df
```

```python
# manually creating dataframe to map team name abbreviations to teams
abbrev_list = [list(pd.read_csv(i).rename(columns={'tm': 'team'})['team'].unique()) for i in glob.glob("../data/PlayerStats/*.csv")]
abbrev_set = set()

for i in abbrev_list:
    abbrev_set |= set(i)
    
abbrev_set 
```

```python
s={
'ARI': 'crd',
'ATL': 'atl',
'BAL': 'rav',
'BUF': 'buf',
'CAR': 'car',
'CHI': 'chi',
'CIN': 'cin',
'CLE': 'cle',
'DAL': 'dal',
'DEN': 'den',
'DET': 'det',
'GNB': 'gnb',
'HOU': 'htx',
'IND': 'clt',
'JAX': 'jax',
'KAN': 'kan',
'LAC': 'sdg',
'LAR': 'ram',
'LVR': 'rai',
'MIA': 'mia',
'MIN': 'min',
'NOR': 'nor',
'NWE': 'nwe',
'NYG': 'nyg',
'NYJ': 'nyj',
'OAK': 'rai',
'PHI': 'phi',
'PIT': 'pit',
'SDG': 'sdg',
'SEA': 'sea',
'SFO': 'sfo',
'STL': 'ram',
'TAM': 'tam',
'TEN': 'oti',
'WAS': 'was',
}

abbrev_df = pd.DataFrame([(i, s[i]) for i in s], columns=["team", 'team_id'])
abbrev_df
```

```python
files = glob.glob("../data/PlayerStats/*.csv")

for i in files:
    print(f"FILE: {i}")
    
    df = pd.read_csv(i)
    
    # use link column to construct player_id column
    df['player_id'] = df['link'].str.split("/", expand=False).str[-1].str.replace(".htm", "")
    
    # drop link column
    df = df.drop(['link'], axis=1)
    
    # renaming columns
    if 'team' not in df.columns and 'tm' in df.columns:
        df = df.rename(columns={'tm': 'team'})
    
    # adding team_id to columns using abbrev_df
    df = df.merge(abbrev_df, on=['team'], how='outer')
    
    # reordering columns
    id_cols = ['player_id', 'team_id', 'name']
    df = df[id_cols + [i for i in df.columns if i not in id_cols]]
    
    print(df)
    print()
    
    df.to_csv(i, index=False)
```
